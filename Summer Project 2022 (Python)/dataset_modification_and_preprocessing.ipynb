### 2.1. Preparing 2021_Clayton Solar PV.xlsx (Timestamp: From 1/1/2021 12:00 AM to 31/12/2021 11:45 PM) <a class="anchor" name="first_df"></a>

 </div>

In this section, the used file is **2021_Clayton Solar PV.xlsx**. The steps to prepare it are as follow:

1. Import pandas library;
2. Read 2021_Clayton Solar PV.xlsx file (i.e., only sheet 1);
3. Remove the first two rows because it does not contain any important information about the PV generation process;
4. Rename the column "0" to "ReadTime";
5. Reset Index because we remove the first two columns in the third step;
6. Drop "Unnamed: 33" column;
7. Reformat ReadTime variable into DD/MM/YYYY HH:MM AM/PM date format (without 0 in the front of day or month such as 01, 02, etc.);
8. Write the final dataset into PV_data_first.csv file and include the observation number.

```python
# Import Library
import pandas as pd
```

```python
# Read .xlsx file
xls = pd.ExcelFile('./data/2021_Clayton Solar PV.xlsx')
PV_df = pd.read_excel(xls, 'Sheet1')
```

```python
# Print the reading result
PV_df
```

```python
# Remove the first two rows
PV_df_final = PV_df.iloc[3:]

# Rename 0 column into ReadTime column by using indexing method
PV_df_final.columns.values[0] = "ReadTime"

# Reset Index
PV_df_final.reset_index(drop=True, inplace=True)

# Drop "Unnamed: 33" column
PV_df_final = PV_df_final.loc[:, ~PV_df_final.columns.isin(['Unnamed: 33'])]

PV_df_final
```

```python
# Reformat ReadTime

# Step 1. Split the date and time into 2 separate variables
PV_df_final[['datePV','timePV']] = PV_df_final["ReadTime"].str.split(" ", 1, expand=True)

# Step 2. Re-format the datePV variable
# Split the datePV variable
PV_df_final[['dayPV','monthPV', 'yearPV']] = PV_df_final['datePV'].str.split('/',expand=True)
# Change monthPV format by deleting front 0 (e.g. 01, 02, etc.)
PV_df_final['monthPV'] = PV_df_final['monthPV'].apply(lambda t: str(int(t)))
# Reassemble the date format of DD/MM/YYYY
PV_df_final["datePV"] = PV_df_final[["dayPV", "monthPV", "yearPV"]].apply("/".join, axis=1)

# Step 3. Re-format the ReadTime variable into DD/MM/YYYY HH:MM AM/PM date format (without 0 in the front of day or month such as 01, 02, etc.)
PV_df_final["ReadTime"] = PV_df_final[["datePV", "timePV"]].apply(" ".join, axis=1)

# Drop the previously created helper variables
PV_df_final = PV_df_final.drop(['datePV', 'timePV', 'yearPV', 'monthPV', 'dayPV'], axis=1)

PV_df_final
```

```python
# Generate csv files
PV_df_final.to_csv('./data/PV_data_first.csv', index = False)
```

### 2.2. Preparing 2020-07 to 2021-06 Clayton PV Data (verified) Pivoted.csv (Timestamp: From 1/7/2020 12:15 AM to 1/7/2021 12:00 AM) <a class="anchor" name="second_df"></a>

 </div>

In this section, the used file is **2020-07 to 2021-06 Clayton PV Data (verified) Pivoted.csv**. The steps to prepare it are as follow:

1. Import pandas library;
2. Read 2020-07 to 2021-06 Clayton PV Data (verified) Pivoted.csv;
3. Create Total Clayton solar connected to main HV NMI variable by summing all the buildings' PVs;
4. Reformat the ReadTime from YYYY-MM-DD HH:MM:SS into DD/MM/YYYY HH:MM AM/PM date format.

```python
# Import Library
import pandas as pd
```

```python
# Read CSV
PV_data = pd.read_csv("./data/2020-07 to 2021-06 Clayton PV Data (verified) Pivoted.csv")
```

```python
# Print the reading result
PV_data
```

```python
# Create Total Clayton solar connected to main HV NMI variable
PV_data['Total Clayton solar connected to main HV NMI']=PV_data.iloc[:,2:len(PV_data.axes[1])].sum(axis=1)

# Re-format the ReadTime so that the 2020-07 to 2021-06 Clayton PV Data (verified) Pivoted.csv and PV_data_first.csv files have 
# a similar date format
# Step 1. Split the date and time into 2 separate variables
PV_data[['datePV','timePV']] = PV_data["ReadTime"].str.split(" ", 1, expand=True)

# Step 2. Re-format the datePV variable
# Split the datePV variable
PV_data[['yearPV','monthPV', 'dayPV']] = PV_data['datePV'].str.split('-',expand=True)
# Change dayPV format by deleting front 0 (e.g. 01, 02, etc.)
PV_data['dayPV'] = PV_data['dayPV'].apply(lambda t: str(int(t)))
# Change monthPV format by deleting front 0 (e.g. 01, 02, etc.)
PV_data['monthPV'] = PV_data['monthPV'].apply(lambda t: str(int(t)))
# Change the format from YYYY-MM-DD into DD/MM/YYYY
PV_data["datePV"] = PV_data[["dayPV", "monthPV", "yearPV"]].apply("/".join, axis=1)

# Step 3. Re-format the timePV variable
# Split the timePV variable
PV_data[['hourPV','minutePV', 'secondPV']] = PV_data['timePV'].str.split(':',expand=True)
# Create PM or AM
PV_data['AMorPM'] = PV_data['hourPV'].apply(lambda t: "AM" if int(t) / 12 < 1 else "PM")
# Change the format from HH:MM:SS into HH:MM AM/PM
PV_data['hourPV'] = PV_data['hourPV'].apply(lambda t: "12" if t == "00" else t)
PV_data['hourPV'] = PV_data['hourPV'].apply(lambda t: str(int(t) - 12) if int(t) > 12 else t)
# Change hourPV format by deleting front 0 (e.g. 01, 02, etc.)
PV_data['hourPV'] = PV_data['hourPV'].apply(lambda t: str(int(t)))
PV_data["timePV"] = PV_data[["hourPV", "minutePV"]].apply(":".join, axis=1)
PV_data["timePV"] = PV_data[["timePV", "AMorPM"]].apply(" ".join, axis=1)

# Step 4. Re-format the ReadTime
PV_data["ReadTime"] = PV_data[["datePV", "timePV"]].apply(" ".join, axis=1)

# Step 5. Exclude previously created helper variables
PV_data = PV_data.drop(['datePV', 'timePV', 'yearPV', 'monthPV', 'dayPV', 'hourPV', 'minutePV', 'secondPV', 'AMorPM'], axis=1)
```

### 2.3. Preparing Weather Dataset (Timestamp: From 31/12/2015 12:15 AM to 16/7/2021 10:45 PM) <a class="anchor" name="4_df"></a>

 </div>

In this section, the used file is **-37.912_145.134_Solcast_PT5M.csv**. The steps to prepare it are as follow:

1. Import pandas library;
2. Read -37.912_145.134_Solcast_PT5M.csv file;
3. Rename PeriodStart into ReadTime (for joining with the final PV dataset by variable name in the upcoming section);
4. Reformat the date from YYYY-MM-DD HH:MM:SS into DD/MM/YYYY HH:MM AM/PM;
5. Filter only minutes 00, 15, 30, and 45;
6. Write the final dataset into weather_data.csv file.

```python
# Import Library
import pandas as pd
```

```python
# Read CSV
weather_data = pd.read_csv("./data/-37.912_145.134_Solcast_PT5M.csv")
```

```python
# Rename PeriodStart into ReadTime -> [We will be using PeriodStart instead of PeriodEnd]
weather_data.rename(columns = {'PeriodStart':'ReadTime'}, inplace = True)

# Re-format the ReadTime so that the final PV dataset and -37.912_145.134_Solcast_PT5M.csv files have 
# a similar date format
# Step 1. Split the date and time into 2 separate variables
weather_data[['datePV','timePV']] = weather_data["ReadTime"].str.split("T", 1, expand=True)

# Step 2. Remove Z from timePV variable
weather_data['timePV'] = weather_data['timePV'].str.replace(r'Z', '')

# Step 3. Re-format the datePV variable
# Split the datePV variable
weather_data[['yearPV','monthPV', 'dayPV']] = weather_data['datePV'].str.split('-',expand=True)
# Change dayPV format by deleting front 0 (e.g. 01, 02, etc.)
weather_data['dayPV'] = weather_data['dayPV'].apply(lambda t: str(int(t)))
# Change monthPV format by deleting front 0 (e.g. 01, 02, etc.)
weather_data['monthPV'] = weather_data['monthPV'].apply(lambda t: str(int(t)))
# Change the format from YYYY-MM-DD into DD/MM/YYYY
weather_data["datePV"] = weather_data[["dayPV", "monthPV", "yearPV"]].apply("/".join, axis=1)

# Step 4. Re-format the timePV variable
# Split the timePV variable
weather_data[['hourPV','minutePV', 'secondPV']] = weather_data['timePV'].str.split(':',expand=True)
# Create PM or AM
weather_data['AMorPM'] = weather_data['hourPV'].apply(lambda t: "AM" if int(t) / 12 < 1 else "PM")
# Change the format from HH:MM:SS into HH:MM AM/PM
weather_data['hourPV'] = weather_data['hourPV'].apply(lambda t: "12" if t == "00" else t)
weather_data['hourPV'] = weather_data['hourPV'].apply(lambda t: str(int(t) - 12) if int(t) > 12 else t)
# Change hourPV format by deleting front 0 (e.g. 01, 02, etc.)
weather_data['hourPV'] = weather_data['hourPV'].apply(lambda t: str(int(t)))
weather_data["timePV"] = weather_data[["hourPV", "minutePV"]].apply(":".join, axis=1)
weather_data["timePV"] = weather_data[["timePV", "AMorPM"]].apply(" ".join, axis=1)

# Step 4. Re-format the ReadTime
weather_data["ReadTime"] = weather_data[["datePV", "timePV"]].apply(" ".join, axis=1)

# Step 5. Filter only minutes 0, 15, 30, and 45
minutes_list = ["00", "15", "30", "45"] # Create minutes_list
weather_data2 = weather_data.loc[weather_data['minutePV'].isin(minutes_list)]
```

```python
# Write CSV of final weather dataset
# Step 1. Exclude previously created helper variables and other unnecessary variables
weather_data2 = weather_data2.drop(['datePV', 'timePV', 'yearPV', 'monthPV', 'dayPV', 'hourPV', 'minutePV', 'secondPV', 'AMorPM', 'PeriodEnd', 'Period'], axis=1)

weather_data2
```

```python
# Step 2. Generate csv files
weather_data2 = weather_data2.reset_index(drop=True) # Reset index

weather_data2.to_csv('./data/weather_data.csv', index = False)
```

#### 3.5.1. Feature Engineering <a class="anchor" name="feature_engineering"></a>

 </div>

**15 Minutes Prior Photovoltaic (PV) Value Feature**

```python
# Convert final_pv1 from wide to long format
# Step 1. Change the format from wide to long
final_pv1 = pd.melt(final_pv1, id_vars='ReadTime',
                    var_name='building_names', value_name='pv_values')

# Step 2. Create 15 minutes prior pv_values variable
final_pv1['pv_values_1'] = final_pv1['pv_values'].shift(1)

# Step 3. Remove missing values
final_pv1 = final_pv1.dropna()

# Step 4. Reset index
final_pv1.reset_index(drop=True, inplace=True)
```

```python
final_pv1
```

**Time, Day, Month, year, and Season Features**

```python
# Step 1. Split the date and time into 2 separate variables
final_pv1[['dateTS','timeTS']] = final_pv1["ReadTime"].str.split(" ", 1, expand=True)

# Step 2. Split the dateTS variable into day, month and year (Create day, month, and year features)
final_pv1[['dayofmonth','month', 'year']] = final_pv1['dateTS'].str.split('/',expand=True)

# Convert day, month, and year into an integer datatype
final_pv1 = final_pv1.astype({"dayofmonth":"int", "month":"int", "year":"int"})

# Step 3. Create season feature
# Ref: http://www.bom.gov.au/climate/glossary/seasons.shtml
summer = [12, 1, 2] # December, January and February
autumn = [3, 4, 5] # March, April and May
winter = [6, 7, 8] # June, July and August
spring = [9, 10, 11] # September, October and November

final_pv1.loc[(final_pv1['month'].isin(summer)), 'seasonTS'] = 0
final_pv1.loc[(final_pv1['month'].isin(autumn)), 'seasonTS'] = 1
final_pv1.loc[(final_pv1['month'].isin(winter)), 'seasonTS'] = 2
final_pv1.loc[(final_pv1['month'].isin(spring)), 'seasonTS'] = 3

# Convert seasonTS into an integer datatype
final_pv1 = final_pv1.astype({"seasonTS":"int"})

# Step 4. Create time feature
# Step 4.1. Split timeTS variable into temptime and PMorAM variables by whitespace
final_pv1[['temptime','AMorPM']] = final_pv1["timeTS"].str.split(" ", 1, expand=True)

# Step 4.2. Split temptime into hourTS and minuteTS variables by semicolon
final_pv1[['hourTS','minutesTS']] = final_pv1["temptime"].str.split(":", 1, expand=True)

# Step 4.3. Convert hourTS and minuteTS into an integer datatype
final_pv1 = final_pv1.astype({"hourTS":"int", "minutesTS":"int"})

# Step 4.4. Convert AMorPM into 24hours
final_pv1.loc[(final_pv1['AMorPM'] == 'AM') & (final_pv1['hourTS'] == 12), 'hourTS'] = int(0) # 12:xx AM into 00:xx
final_pv1['hourTS'] = final_pv1.apply(lambda x: x.hourTS + int(12) if x.AMorPM == "PM" else x.hourTS, axis = 1) # xx:xx PM into 24hours

# Step 4.5. Create timeTS by calculating how many minutes have passed from 00:00 (e.g., 01:15 will be 75)
final_pv1['timeTS'] = final_pv1['hourTS'] * int(60) + final_pv1['minutesTS']

# Step 5. Drop the helper variables
final_pv1 = final_pv1.drop(['dateTS', 'temptime', 'AMorPM', 'hourTS', 'minutesTS'], axis=1)
```

```python
final_pv1
```

**Building / Location Dummy Features**

- **time_series_df.csv:** the combination of the PV_data.csv and weather_data.csv files. This file has a timestamp from 1/7/2020 12:15 AM to 16/7/2021 10:45 PM and is used for the RShiny App (i.e., Time Series Data Analytics and Visualization App) because the dataset still contains the missing value and other anomalies;
